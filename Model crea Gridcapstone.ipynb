```python
import pandas as pd
```

```python
dataset=pd.read_csv("Crop_recommendation.csv",index_col=None)
```

```python
df=dataset
```

```python
df
```

```python
df=pd.get_dummies(df,dtype=int,drop_first=True)
```

```python
df
```

```python
df.columns
```

```python
independent=df[['N','P','K','temperature','humidity','ph','rainfall']]
```

```python
independent
```

```python
dependent=df[['label_banana', 'label_blackgram', 'label_chickpea', 'label_coconut',
       'label_coffee', 'label_cotton', 'label_grapes', 'label_jute',
       'label_kidneybeans', 'label_lentil', 'label_maize', 'label_mango',
       'label_mothbeans', 'label_mungbean', 'label_muskmelon', 'label_orange',
       'label_papaya', 'label_pigeonpeas', 'label_pomegranate', 'label_rice',
       'label_watermelon']]
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50, random_state = 0)
regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_Forest.sav"
pickle.dump(regressor,open(filename,'wb'))
```
